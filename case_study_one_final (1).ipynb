Function to calculate IQR and lower and upper bound of IQR

```python
def lower_higher(column):
    lower = df[column].quantile(.25)
    upper = df[column].quantile(.75)
    median = df[column].quantile(.50)
    IQR = upper-lower
    lowest = round(lower - 1.5*IQR,2)
    highest = round(upper + 1.5*IQR ,2)
    return (lowest,highest,median)
```

```python
print(df['MedianAge'].skew())
```

```python
low,high,median = lower_higher('MedianAge')
print(low,high,median)
```
